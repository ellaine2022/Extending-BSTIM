## Explore borreliosis dataset

```python
import pandas as pd 

df_borreliosis = pd.read_csv("../original_data_file/diseases/borreliosis.csv", delimiter=';')
```

```python
df_borreliosis.head()
```

```python
df_borreliosis.tail()
```

```python
df_borreliosis.shape
```

```python
print(df_borreliosis.isnull().sum())
```

```python
print("yes" if df_borreliosis.isnull().values.any() else "no")
```

```python
df_borreliosis.columns
```

```python
# Extract the year part from the 'day' column
df_borreliosis['year'] = df_borreliosis['day'].str.extract(r'^(\d{4})')

# Show unique years
print(df_borreliosis['year'].unique())
```

```python
# Count rows for each year
year_counts = df_borreliosis['year'].value_counts().sort_index()

print(year_counts)
```

```python
df_borreliosis.head()
```

```python
df_borreliosis_filtered = df_borreliosis[df_borreliosis['year'].astype(int).between(2013, 2017)]
```

```python
df_borreliosis_filtered.shape
```

```python
# Count rows for each year
year_counts_filtered = df_borreliosis_filtered['year'].value_counts().sort_index()

print(year_counts_filtered)
```

```python
df_borreliosis_filtered.columns
```

```python
df_borreliosis_filtered.to_csv('../data/diseases/borreliosis.csv', index=False)
```
